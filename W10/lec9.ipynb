```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal
from scipy.signal import freqz
```

```python
def compute_module(a):
    return np.sqrt(a.imag**2 + a.real**2)
```

## Fig 1: Input Signal

```python
t = np.linspace(0, 1, 10000)
x = 4 + np.cos(250 * np.pi * t - np.pi/4) - 3 * np.cos(2000 * np.pi * t / 3)
```

```python
fig, ax = plt.subplots(figsize=(20, 4))
ax.plot(t, x, color='b')
ax.set_xlim(0, 1.01)
ax.set_ylabel("Amplitude")
ax.set_xlabel('Time in second')
ax.grid(True)
ax.set_title('Input Signal')
```

## Fig 2: Magnitude Spectrum of Input Signal

```python
t = np.linspace(0, 1, 1000)
x = 4 + np.cos(250 * np.pi * t - np.pi/4) - 3 * np.cos(2000 * np.pi * t / 3)
sp = np.fft.fft(x)
freq = np.fft.fftfreq(t.shape[-1])
w = freq * 2 * np.pi
sp_m = compute_module(sp)
```

```python
fig, ax = plt.subplots(figsize=(20, 4))
ax.plot(w, sp_m)
ax.set_xlim(0, 3.1)
ax.set_ylabel("Magnitude")
ax.set_xlabel('Frequency in rad/s')
ax.grid(True)
ax.set_title('Magnitude Spectrum of Input Signal')
```

## Fig 3: Magnitude Response of Filter

```python
t = np.linspace(0, 1, 1000)
x = 4 + np.cos(250 * np.pi * t - np.pi/4) - 3 * np.cos(2000 * np.pi * t / 3)
b = [1, 1, 1]
a = [3, 0, 0]
w, h = signal.freqz(b, a)
```

```python
magnitude = compute_module(h)
fig, ax = plt.subplots(figsize=(20, 4))
ax.plot(w, magnitude, color='b')
ax.set_ylabel("Magnitude")
ax.set_xlabel('Frequency in rad/s')
ax.grid(True)
ax.set_title('Magnitude Response of Filter')
```

## Fig 4: Output Signal

```python
t = np.linspace(0, 1, 1000)
x = 4 + np.cos(250 * np.pi * t - np.pi/4) - 3 * np.cos(2000 * np.pi * t / 3)
y = signal.lfilter(b, a, x)
```

```python
fig, ax = plt.subplots(figsize=(20, 4))
ax.plot(t, y, color='b')
ax.set_xlim(0, 1.01)
ax.set_ylim(3.1, 5)
ax.set_ylabel("Amplitude")
ax.set_xlabel('Time in second')
ax.grid(True)
ax.set_title('Output Signal')
```

## Fig 5: Magnitude Spectrum of Output Signal

```python
t = np.linspace(0, 1, 1000)
sp = np.fft.fft(y)
freq = np.fft.fftfreq(t.shape[-1])
w = freq * 2 * np.pi
sp_m = compute_module(sp)
```

```python
fig, ax = plt.subplots(figsize=(20, 4))
ax.plot(w, sp_m)
ax.set_xlim(0, 3.1)
ax.set_ylabel("Magnitude")
ax.set_xlabel('Frequency in rad/s')
ax.grid(True)
ax.set_title('Magnitude Spectrum of Output Signal')
```
